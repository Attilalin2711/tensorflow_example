```python
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
image_raw_data_jpg = tf.gfile.FastGFile('C://Users//User//Desktop//tf_example//00478403_20160401_CT_2_1_1.jpg','rb').read()
#‘r’:UTF-8編碼; ‘rb’:非UTF-8編碼 ,r會有錯誤
```

```python
with tf.Session() as sess:
    img_data_jpg_decode = tf.image.decode_jpeg(image_raw_data_jpg)#圖像解碼
    img_data_jpg_convert = tf.image.convert_image_dtype(img_data_jpg_decode, dtype=tf.uint8)
    plt.figure(1)
    print('decode',sess.run(img_data_jpg_decode))
    print('convert',sess.run(img_data_jpg_convert))#顯示影像矩陣
    plt.imshow(img_data_jpg_convert.eval())
    plt.show()
```
